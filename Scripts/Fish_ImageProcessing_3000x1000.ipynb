```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
import imageio
from PIL import Image

# # Set the directory path of caeruleum fish images
path = 'blabla/fish_images/caeruleum/JPEG/male/'

files = os.listdir(path)
n_files = len(files)
print(n_files)

FISH patch_dir = 'blabla/fish_images/caeruleum/patches_3000x1000/'
```

```python
for im_ndx in files: 
    print('image name: {}'.format(im_ndx))
    image    = im_ndx
    img_path = os.path.join(path, image)
    image = Image.open(img_path)
   
    # Display the original image
    plt.figure()
    plt.imshow(image) 
    plt.show()     

    # Setting the points for the cropped image 
    left = 1700 #to adapt: fish are not centred the same way on the different images
    right = left + 3000
    top = 1700  #to adapt: fish are not centred the same way on the different images
    bottom = top + 1000
  
    # Cropped image of above dimension 
    # (It will not change the original image) 
    im1 = image.crop((left, top, right, bottom)) 
  
    # Display the cropped image
    plt.figure()
    plt.imshow(im1) 
    plt.show()
    
    print(im1.size)
    
#     im1.save(patch_dir + 'cropped_3000x1000_'+ im_ndx[0:-4] + '.jpg')    
#     plt.imsave(patch_dir + 'cropped_3000x1000_'+ im_ndx[0:-4] + '.jpg', im1)

```
